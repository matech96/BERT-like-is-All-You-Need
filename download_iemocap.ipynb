```python
import subprocess
from multiprocessing import Pool
from functools import partial
from tqdm import tqdm
```

```python
with open("../T_data/iemocap/train_t.tsv") as f:
    lines = f.readlines()
base = [l[:-5] for l in lines]
```

```python
def f(b, folder_id, ext):
    subprocess.run(["../gdown.pl", f"https://drive.google.com/drive/folders/{folder_id}", f"{b}.{ext}"])
    
    
def download(folder_id, ext):
        
    with Pool(50) as p:
        r = list(tqdm(p.imap(partial(f, folder_id=folder_id, ext=ext), base), total=len(base)))
#         for b in :
#             subprocess.run(["../gdown.pl", f"https://drive.google.com/drive/folders/{folder_id}", f"{b}.{ext}"])
    return r
```
